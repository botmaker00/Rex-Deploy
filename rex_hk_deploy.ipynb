# HI: Upar jo folders config kiye the, unka naam yahan likho.

for app in Deploy_List.split():
  app = app.strip()
  if not app:

    continue
    
  # Verify config.env exists before deploying
  config_path = os.path.join(app, "config.env")
  if not os.path.exists(config_path):
    print(f"‚ùå {app}/config.env NOT found! Skipping deployment.")

  %cd {app}
  
  !git add . -f
  !git commit -m "Rex Bots Deploy" --allow-empty
  !heroku git:remote -a {app}
  !git push heroku main -f

  print(f"   The bot will load settings from config.env at runtime.\n")
  
  %cd ..

print("\nüéâ All deployments completed!")


```python
# ================================================
# üîß OPTIONAL: Set Heroku Config Vars from File
# ================================================
# EN: By default, the bot reads config.env from the deployed files.
#     But if you prefer to use Heroku Config Vars instead, run this cell.
# HI: Default me bot config.env file se hi settings padhta hai.
#     Agar tum Heroku Config Vars use karna chahte ho, to ye cell run karo.

App_Name_For_Vars = ""  # @param {type:"string"}

import os

if not App_Name_For_Vars.strip():
  print("‚ö†Ô∏è Please enter App_Name_For_Vars to continue.")
else:
  config_path = os.path.join(App_Name_For_Vars, "config.env")
  
  if not os.path.exists(config_path):
    print(f"‚ùå {config_path} not found!")
  else:
    # Read config.env and parse it
    with open(config_path, 'r') as f:
      lines = f.readlines()
    
    config_vars = []
    for line in lines:
      line = line.strip()
      if not line or line.startswith('#'):
        continue
      if '=' in line:
        key, value = line.split('=', 1)
        key = key.strip()
        value = value.strip().strip('"').strip("'")
        if key and value:
          config_vars.append(f'{key}={value}')
    
    if config_vars:
      # Set all config vars at once
      vars_str = ' '.join(config_vars)
      !heroku config:set {vars_str} -a {App_Name_For_Vars}
      print(f"\n‚úÖ Config vars set for {App_Name_For_Vars}")
      print("   Note: The bot code uses load_dotenv('config.env'), so the")
      print("   deployed config.env file will still be checked first.")
    else:
      print("‚ùå No valid config variables found in config.env")
```

üí° **When to use this cell:**
- **English:** Only run this if you want environment variables in the Heroku dashboard in addition to the `config.env` file. The bot primarily loads from `config.env`, but Config Vars can serve as a backup or override.
- **Hindi:** Agar tum chahte ho ki Heroku dashboard me bhi variables dikhen, tabhi ye run karo. Bot pehle `config.env` se load karta hai, lekin Config Vars backup ya override ke kaam aa sakte hain.


```python
# ===============================
# 4Ô∏è‚É£ Deploy All Configured Apps
# ===============================
# EN: Write the app folder names you configured above.
# HI: Upar jo folders config kiye the, unka naam yahan likho.

Deploy_List = ""  # @param {type:"string"}

import os

for app in Deploy_List.split():
  app = app.strip()
  if os.path.isdir(app):
    %cd {app}
    !git add . -f
    !git commit -m "Rex Bots Deploy" --allow-empty
    !heroku git:remote -a {app}
    !git push heroku main -f
    print(f"‚úÖ {app} ‚Üí Deployed Successfully!")
    %cd ..
  else:
    print(f"‚ö†Ô∏è {app} folder not found!")
```
