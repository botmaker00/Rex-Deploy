```
# 1. HEROKU LOGIN (Zaroori Hai Bhai)

Heroku_Email = "" # @param {type:"string"}
Heroku_API_Key = "" # @param {type:"string"}

!curl -s https://cli-assets.heroku.com/install.sh | sh
from IPython.display import clear_output, HTML; clear_output()

if not Heroku_Email or not Heroku_API_Key:
  raise Exception("Heroku Email aur API Key daal na bhai! Nahi toh deploy nahi hoga!")

import os
with open(os.path.expanduser("~/.netrc"), "w") as f:
  f.write(f"machine api.heroku.com\n  login {Heroku_Email}\n  password {Heroku_API_Key}\n"
          f"machine git.heroku.com\n  login {Heroku_Email}\n  password {Heroku_API_Key}")
os.chmod(os.path.expanduser("~/.netrc"), 0o600)

!git config --global user.email {Heroku_Email}
!git config --global user.name "RexKing"

display(HTML("""
<div style='text-align:center; padding:20px; background:linear-gradient(45deg,#00c853,#64dd17); color:white; border-radius:20px; font-size:22px;'>
   HEROKU LOGIN SUCCESSFUL!<br>
   <b>Ab full power mode on hai bhai!</b>
</div>"""))
```

```
# 2. BOT DETAILS (Yahin daal do, seedha Heroku pe set ho jayega)

API_ID = 12345678 # @param {type:"integer"}
API_HASH = "" # @param {type:"string"}
BOT_TOKEN = "" # @param {type:"string"}
OWNER_ID = 1234567890 # @param {type:"integer"}
SUDO_USERS = "" # @param {type:"string"}  # IDs space se alag: 123 456 789
LOG_CHANNEL = -1001234567890 # @param {type:"integer"}
SESSION_NAME = "RexEncoderX" # @param {type:"string"}
MONGO_URI = "" # @param {type:"string"}
UPSTREAM_REPO = "https://github.com/abhinai2244/Encoding-Bot" # @param {type:"string"}
UPSTREAM_BRANCH = "main" # @param {type:"string"}

# Optional (chhod sakte ho)
INDEX_URL = "" # @param {type:"string"}
DRIVE_DIR = "" # @param {type:"string"}
DOWNLOAD_DIR = "VideoEncoder/downloads/" # @param {type:"string"}
ENCODE_DIR = "VideoEncoder/encodes/" # @param {type:"string"}

display(HTML("""
<div style='text-align:center; padding:25px; background:#311b92; color:#e8f5e8; border-radius:20px; border:3px solid #7c4dff;'>
   <h2>Sab Details Ready!</h2>
   <h3 style='color:#76ff03;'>Ab sirf deploy karo aur bot live!</h3>
</div>"""))
```

```
# 3. NAYI HEROKU APPS BANAYE (Optional)

Create_New_Apps = "rex-vip1 rex-pro2 rex-king3" # @param {type:"string"}
Region = "us" # @param ["us", "eu"]

for app in Create_New_Apps.split():
  app = app.strip()
  if app:
    !heroku create {app} --region {Region} --stack container
    print(f"{app} BAN GAYA BHAI!")
```

```
# 4. FULL POWER DEPLOY (Sab apps ek saath!)

Deploy_These_Apps = "rex-vip1 rex-pro2 rex-king3" # @param {type:"string"}

config_vars = {
    "API_ID": str(API_ID),
    "API_HASH": API_HASH,
    "BOT_TOKEN": BOT_TOKEN,
    "OWNER_ID": str(OWNER_ID),
    "SUDO_USERS": SUDO_USERS,
    "LOG_CHANNEL": str(LOG_CHANNEL),
    "SESSION_NAME": SESSION_NAME,
    "MONGO_URI": MONGO_URI,
    "UPSTREAM_REPO": UPSTREAM_REPO,
    "UPSTREAM_BRANCH": UPSTREAM_BRANCH,
    "INDEX_URL": INDEX_URL,
    "DRIVE_DIR": DRIVE_DIR,
    "DOWNLOAD_DIR": DOWNLOAD_DIR,
    "ENCODE_DIR": ENCODE_DIR
}

for app in Deploy_These_Apps.split():
  app = app.strip()
  if not app: continue
  print(f"\nDeploying {app}...")
  
  !rm -rf {app}
  !git clone https://github.com/botmaker00/Rex-Deploy {app}
  %cd {app}
  !rm -f README.md *.ipynb
  
  !heroku git:remote -a {app}
  
  for key, val in config_vars.items():
    if val and str(val).strip():
      !heroku config:set {key}="{val}" -a {app}
  
  !git add . -f
  !git commit -m "Rex Full Power Deploy" --allow-empty
  !git push heroku main -f
  
  print(f"{app} LIVE HO GAYA BHAI!")
  %cd ..

display(HTML("""
<div style='text-align:center; padding:30px; background:linear-gradient(45deg,#ff1744,#f50057); color:white; border-radius:30px; font-size:28px; margin:20px;'>
   SAB BOTS LIVE HO GAYE!<br>
   <b>REX OP HAI!</b>
</div>"""))
```
