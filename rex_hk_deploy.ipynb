```python
#@title 3. NAYI APPS BANAYE (Optional)

Create_New_Apps = "rex-op1 rex-op2 rex-king" #@param {type:"string"}
Region = "us" #@param ["us", "eu"]

for app in Create_New_Apps.split():
  app = app.strip()
  if app:
    !heroku create {app} --region {Region} --stack container
    print(f"{app} BAN GAYA BHAI!")
```

```python
#@title 4. FULL POWER DEPLOY (Sab ek saath!)

Deploy_These_Apps = "rex-op1 rex-op2 rex-king" #@param {type:"string"}

config_vars = {
    "API_ID": str(API_ID),
    "API_HASH": API_HASH,
    "BOT_TOKEN": BOT_TOKEN,
    "OWNER_ID": str(OWNER_ID),
    "SUDO_USERS": SUDO_USERS,
    "LOG_CHANNEL": str(LOG_CHANNEL),
    "SESSION_NAME": SESSION_NAME,
    "MONGO_URI": MONGO_URI,
    "UPSTREAM_REPO": UPSTREAM_REPO,
    "UPSTREAM_BRANCH": UPSTREAM_BRANCH,
    "INDEX_URL": INDEX_URL,
    "DRIVE_DIR": DRIVE_DIR,
    "DOWNLOAD_DIR": DOWNLOAD_DIR,
    "ENCODE_DIR": ENCODE_DIR
}

for app in Deploy_These_Apps.split():
  app = app.strip()
  if not app: continue
  print(f"\nDeploying {app}...")
  
  !rm -rf {app}
  !git clone https://github.com/botmaker00/Rex-Deploy {app}
  %cd {app}
  !rm -f README.md *.ipynb
  
  !heroku git:remote -a {app}
  
  for key, val in config_vars.items():
    if val and str(val).strip():
      !heroku config:set {key}="{val}" -OW -a {app}
  
  !git add . -f
  !git commit -m "Rex OP Deploy" --allow-empty
  !git push heroku main -f
  
  print(f"{app} LIVE HO GAYA BHAI!")
  %cd ..

display(HTML("""
<div style='text-align:center; padding:40px; background:linear-gradient(45deg,#ff1744,#c62828); color:white; border-radius:40px; font-size:32px; margin:30px;'>
   SAB BOTS LIVE!<br><br>
   <b>REX HAI TOH MUMKIN HAI!</b>
</div>"""))
```
