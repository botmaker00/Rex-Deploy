```python
#@title 2Ô∏è‚É£ Create Heroku Apps (Optional)
# EN: Enter app names (space-separated). Skip if apps already exist. (Optional: Enter Team Name if needed)
# HI: App names likho (space se alag). Agar pehle se bani hain to skip karo. (Optional: Agar Team hai to naam likho)

App_Names = "" #@param {type:"string"}
Region = "eu" #@param ["eu", "us"]
Team_Name = "" #@param {type:"string"}

team_flag = f"--team {Team_Name}" if Team_Name.strip() else ""
apps_created = []
for name in App_Names.split():
  clean_name = name.strip().lower().replace(" ", "-"[:1])
  if clean_name:
    !heroku create {clean_name} --region {Region} --stack container {team_flag}
    print(f"‚úÖ Created: {clean_name}")
    apps_created.append(clean_name)

if apps_created:
  print(f"\nüìù Your apps: { ' '.join(apps_created) }")
```

```python
#@title 3Ô∏è‚É£ Config Setup + Heroku Push 
# EN: Choose method, fill details, and it will push to Heroku safely.
# HI: Method choose karo, details bharo, Heroku pe safe push ho jayega.

App_Name = "" #@param {type:"string"}
Config_Method = "Manual" #@param ["Manual", "Gist URL", "Upload File"]

import os
import shutil
import subprocess
from IPython.display import clear_output

# Clean old folder if exists
if os.path.exists(App_Name):
  shutil.rmtree(App_Name)

# Clone repo
!git clone https://github.com/botmaker00/Rex-Deploy "{App_Name}"
%cd "{App_Name}"

# Remove extra files
!rm -f README.md *.ipynb *.sh 2>/dev/null || true

# FIXED FUNCTION: Safe Heroku config set (no shell quotes issue)
def set_heroku_var(key, value):
  try:
    value_str = str(value)
    # Use subprocess to avoid shell quoting problems
    cmd = ["heroku", "config:set", f"{key}={value_str}", "-a", App_Name]
    result = subprocess.run(cmd, capture_output=True, text=True, timeout=30)
    if result.returncode == 0:
      print(f"‚úÖ {key} = {value_str[:20]}... (Set successfully)")
      return True
    else:
      print(f"‚ùå {key} failed: {result.stderr.strip()}")
      return False
  except Exception as e:
    print(f"‚ùå {key} error: {str(e)}")
    return False

print("üîÑ Setting up config...")

if Config_Method == "Manual":
  # Manual inputs
  API_ID = None #@param {type:"integer"}
  API_HASH = "" #@param {type:"string"}
  BOT_TOKEN = "" #@param {type:"string"}
  OWNER_ID = None #@param {type:"integer"}
  SUDO_USERS = "" #@param {type:"string"}
  EVERYONE_CHATS = "" #@param {type:"string"}
  LOG_CHANNEL = None #@param {type:"integer"}
  SESSION_NAME = "rexencoder" #@param {type:"string"}
  MONGO_URI = "" #@param {type:"string"}
  UPSTREAM_REPO = "https://github.com/abhinai2244/Encoding-Bot" #@param {type:"string"}
  UPSTREAM_BRANCH = "main" #@param {type:"string"}

  if not all([API_ID, API_HASH, BOT_TOKEN, OWNER_ID]):
    raise Exception("‚ùå Fill API_ID, API_HASH, BOT_TOKEN, OWNER_ID first!")

  # Create config.env for bot
  config_text = f"""API_ID={API_ID}
API_HASH={API_HASH}
BOT_TOKEN={BOT_TOKEN}
OWNER_ID={OWNER_ID}
SUDO_USERS={SUDO_USERS}
EVERYONE_CHATS={EVERYONE_CHATS}
LOG_CHANNEL={LOG_CHANNEL}
SESSION_NAME={SESSION_NAME}
MONGO_URI={MONGO_URI}
UPSTREAM_REPO={UPSTREAM_REPO}
UPSTREAM_BRANCH={UPSTREAM_BRANCH}"""

  with open("config.env", "w") as f:
    f.write(config_text)
  print("üìù config.env created from manual input")

  # Push each var safely to Heroku
  vars_set = 0
  set_heroku_var("API_ID", API_ID)
  set_heroku_var("API_HASH", API_HASH)
  set_heroku_var("BOT_TOKEN", BOT_TOKEN)
  set_heroku_var("OWNER_ID", OWNER_ID)
  set_heroku_var("SUDO_USERS", SUDO_USERS)
  set_heroku_var("EVERYONE_CHATS", EVERYONE_CHATS)
  set_heroku_var("LOG_CHANNEL", LOG_CHANNEL)
  set_heroku_var("SESSION_NAME", SESSION_NAME)
  set_heroku_var("MONGO_URI", MONGO_URI)
  set_heroku_var("UPSTREAM_REPO", UPSTREAM_REPO)
  set_heroku_var("UPSTREAM_BRANCH", UPSTREAM_BRANCH)
  print(f"\nüéâ {vars_set} vars pushed to Heroku!")

elif Config_Method == "Gist URL":
  Gist_URL = "" #@param {type:"string"}
  if not Gist_URL:
    raise Exception("‚ùå Enter Gist RAW URL!")
  !curl -L -o config.env "{Gist_URL}" > /dev/null
  print("üåê config.env downloaded from Gist")

  # Parse and push
  vars_set = 0
  with open("config.env", "r") as f:
    for line in f:
      line = line.strip()
      if '=' in line and not line.startswith('#'):
        key, val = line.split('=', 1)
        key = key.strip()
        val = val.strip().strip('"').strip("'")
        if set_heroku_var(key, val):
          vars_set += 1
  print(f"\nüéâ {vars_set} vars from Gist pushed!")

elif Config_Method == "Upload File":
  from google.colab import files
  uploaded = files.upload()
  if "config.env" not in uploaded:
    raise Exception("‚ùå Upload config.env file!")
  print("üìÅ config.env uploaded")

  # Parse and push (same as Gist)
  vars_set = 0
  with open("config.env", "r") as f:
    for line in f:
      line = line.strip()
      if '=' in line and not line.startswith('#'):
        key, val = line.split('=', 1)
        key = key.strip()
        val = val.strip().strip('"').strip("'")
        if set_heroku_var(key, val):
          vars_set += 1
  print(f"\nüéâ {vars_set} vars from file pushed!")

%cd ..
print("\nüìÇ Configured folders:")
!ls -d */ 2>/dev/null || echo "No folders yet"
```

‚≠ê **Choose Manual** for new setup. **Gist URL** for shared config. **Upload** for local file.

**Example Gist URL:** `https://gist.githubusercontent.com/username/abc123/raw/config.env`

```python
#@title 4Ô∏è‚É£ Deploy Apps
# EN: Enter folder names to deploy (space-separated).
# HI: Deploy karne ke liye folder names likho (space se alag).

Deploy_List = "" #@param {type:"string"}

if not Deploy_List.strip():
  raise Exception("‚ùå Enter app/folder names to deploy!")

import os
deployed = 0
for app in Deploy_List.split():
  app = app.strip()
  if os.path.isdir(app):
    %cd {app}
    !git init 2>/dev/null || true
    !git add . -f
    !git commit -m "Rex Encoder Deploy - $(date)" --allow-empty
    !heroku git:remote -a {app}
    !git push heroku main -f
    print(f"üéâ {app} Deployed Successfully!")
    %cd ..
    deployed += 1
  else:
    print(f"‚ö†Ô∏è Folder '{app}' not found! Create config first.")

print(f"\n‚úÖ Total deployed: {deployed}")
```

üìå **Example:** `rex-encoder-1 rex-encoder-2` (use exact folder names from step 3).
