```
#@title <font color="#FF0000">üîê Step 1: Heroku Login (Sabse Pehle Ye Karo!)</font>
Heroku_Email = "" #@param {type:"string"}
Heroku_API_Key = "" #@param {type:"string"}

from IPython.display import clear_output, HTML
import os

!curl -s https://cli-assets.heroku.com/install.sh | sh
clear_output()

if not Heroku_Email or not Heroku_API_Key:
    display(HTML('<h3><font color="red">‚ùå Heroku Email aur API Key daalo bhai! Nahi to aage nahi badhega!</font></h3>'))
else:
    netrc_content = f"machine api.heroku.com\n  login {Heroku_Email}\n  password {Heroku_API_Key}\nmachine git.heroku.com\n  login {Heroku_Email}\n  password {Heroku_API_Key}"
    with open(os.path.expanduser("~/.netrc"), "w") as f:
        f.write(netrc_content)
    os.chmod(os.path.expanduser("~/.netrc"), 0o600)

    !git config --global user.email {Heroku_Email}
    !git config --global user.name "RexBot_User"

    display(HTML('<h2><font color="#00FF00">‚úÖ Heroku Login Successful! Ab aage badho üòé</font></h2>'))
```

```
#@title <font color="#00FFFF">üöÄ Step 2: Naya Heroku App Banaye? (Optional)</font>
App_Names = "" #@param {type:"string"}
Region = "eu" #@param ["eu", "us"] {allow-input: false}
Team_Name = "" #@param {type:"string"}

if App_Names.strip():
    team_flag = f"--team {Team_Name}" if Team_Name else ""
    for name in App_Names.split():
        name = name.strip().lower()
        if name:
            !heroku create {name} --region {Region} --stack container {team_flag}
            print(f"‚úÖ App bana diya: https://{name}.herokuapp.com")
else:
    print("‚ÑπÔ∏è Koi app name nahi dala, purana app use kar sakte ho.")
```

```
#@title <font color="#FF6600">‚öôÔ∏è Step 3: Config Setup (Sabse Important Step!)</font>
App_Name = "rexencoder" #@param {type:"string"}

#@markdown ### <font color="#00FF00">Method 1: Manual Fill (Sabse Easy)</font>
API_ID = 12345678 #@param {type:"integer"}
API_HASH = "" #@param {type:"string"}
BOT_TOKEN = "" #@param {type:"string"}
OWNER_ID = 1234567890 #@param {type:"integer"}
SUDO_USERS = "" #@param {type:"string"}
LOG_CHANNEL = -1001234567890 #@param {type:"integer"}
SESSION_NAME = "rexencoder" #@param {type:"string"}
MONGO_URI = "" #@param {type:"string"}

#@markdown ### <font color="#FF00FF">Method 2: Gist Raw URL Daalo</font>
Gist_URL = "" #@param {type:"string"}

#@markdown ### <font color="#FFFF00">Method 3: config.env Upload Karo</font>
Upload_Config_File = False #@param {type:"boolean"}

import os, shutil
from IPython.display import clear_output, HTML

if os.path.exists(App_Name):
    shutil.rmtree(App_Name)

!git clone https://github.com/botmaker00/Rex-Deploy {App_Name}
%cd {App_Name}
!rm -f README.md *.ipynb 2>/dev/null || true

if Upload_Config_File:
    from google.colab import files
    uploaded = files.upload()
    for fn in uploaded.keys():
        os.rename(fn, "config.env")
    clear_output()
    display(HTML('<h3><font color="lime">‚úÖ config.env upload ho gaya!</font></h3>'))

elif Gist_URL.strip():
    !curl -L -o config.env "{Gist_URL}"
    clear_output()
    display(HTML('<h3><font color="cyan">‚úÖ Gist se config.env download ho gaya!</font></h3>'))

else:
    config_text = f'''# REX ENCODER BOT CONFIG
API_ID = {API_ID}
API_HASH = "{API_HASH}"
BOT_TOKEN = "{BOT_TOKEN}"
OWNER_ID = {OWNER_ID}
SUDO_USERS = {SUDO_USERS}
LOG_CHANNEL = {LOG_CHANNEL}
SESSION_NAME = "{SESSION_NAME}"
MONGO_URI = "{MONGO_URI}"
UPSTREAM_REPO = "https://github.com/abhinai2244/Encoding-Bot"
UPSTREAM_BRANCH = "main"
DOWNLOAD_DIR = "VideoEncoder/downloads/"
ENCODE_DIR = "VideoEncoder/encodes/"
'''
    with open("config.env", "w") as f:
        f.write(config_text)
    clear_output()
    display(HTML('<h3><font color="gold">‚úÖ Manual config.env ban gaya!</font></h3>'))

%cd ..
display(HTML(f'<h3>üéâ <b>{App_Name}</b> ready for deploy!</h3>'))
```

```
#@title <font color="#FF0066">üöÄ Step 4: Sab Apps Deploy Karo!</font>
Deploy_These_Apps = "rexencoder" #@param {type:"string"}

from IPython.display import clear_output, HTML

success = 0
for app in [x.strip() for x in Deploy_These_Apps.split() if x.strip()]:
    if os.path.isdir(app):
        %cd {app}
        !git add . -f
        !git commit -m "Rex Deploy - $(date)" --allow-empty > /dev/null 2>&1
        !heroku git:remote -a {app} > /dev/null 2>&1 || echo "Remote already exists"
        !git push heroku main -f
        print(f"\nüéâ <b>{app}</b> ‚Üí <a href='https://{app}.herokuapp.com'>LIVE HO GAYA!</a>")
        success += 1
        %cd ..
    else:
        print(f"‚ùå Folder nahi mila: {app}")

clear_output()
if success > 0:
    display(HTML(f'<h1><font color="lime">üéä {success} Bot(s) Successfully Deployed! Enjoy Boss üî•</font></h1>'))
else:
    display(HTML('<h3><font color="red">‚ùå Koi bhi app deploy nahi hua!</font></h3>'))
```

```
#@title <font color="#00FFAA">üìä Step 5: Bot Logs Dekho (Real Time)</font>
App_Name_For_Logs = "" #@param {type:"string"}

if App_Name_For_Logs.strip():
    print(f"üî¥ Live logs for {App_Name_For_Logs} (Ctrl+C to stop)\n")
    !heroku logs -t -a {App_Name_For_Logs}
else:
    print("‚ö†Ô∏è App ka naam daalo bhai!")
```
